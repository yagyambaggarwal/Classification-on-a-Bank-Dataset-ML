```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
```

```python
data = pd.read_csv("./Resources/train.csv")
data.head(10)
```

```python
data.info()
```

```python
for name, content in data.items():
  if not pd.api.types.is_numeric_dtype(content):
    print(name)
```

```python
for name, content in data.items():
  if not pd.api.types.is_numeric_dtype(content):
    data[name] = pd.Categorical(content).codes

data.info()      
```

```python
data.drop("id", axis= 1, inplace = True)
data.info()
```

```python
data = data.sample(frac = 1)
X = data.drop("y", axis = 1)
y = data["y"]

X.info(), y.info()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.6, random_state = 42)
```

```python
model = RandomForestClassifier(n_estimators=3000, n_jobs = -1, max_depth = 30, max_features = "sqrt")
model.fit(X_train, y_train)
# y_pred_proba = model.predict_proba(X_test)
model.score(X_test, y_test)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

```python
model = RandomForestClassifier(n_estimators = 5000, n_jobs = -1, max_depth = 40, max_features = "sqrt")
model.fit(X_train, y_train)
model.score(X_test, y_test)
```
